```python
def analyze_segment(segment, column_name):

    mean = segment[column_name].mean()
    std = segment[column_name].std()

    lower_limit = mean - std
    upper_limit = mean + std

    normal = segment[
        (segment[column_name] >= lower_limit) &
        (segment[column_name] <= upper_limit)
    ]

    abnormal = segment[
        (segment[column_name] < lower_limit) |
        (segment[column_name] > upper_limit)
    ]

    normal_stats = {
        "Mean": normal[column_name].mean(),
        "Std": normal[column_name].std(),
        "Variance": normal[column_name].var(),
        "Min": normal[column_name].min(),
        "Max": normal[column_name].max(),
        "Count": len(normal)
    }

    abnormal_stats = {
        "Mean": abnormal[column_name].mean(),
        "Std": abnormal[column_name].std(),
        "Variance": abnormal[column_name].var(),
        "Min": abnormal[column_name].min(),
        "Max": abnormal[column_name].max(),
        "Count": len(abnormal)
    }

    print("\n==============================")
    print(column_name)
    print("==============================")

    print("\nNORMAL")
    print(normal_stats)

    print("\nABNORMAL")
    print(abnormal_stats)

    return normal, abnormal
```

```python
import pandas as pd

def segment_stats(segment, segment_name, column_name):

    mean = segment[column_name].mean()
    std = segment[column_name].std()

    lower = mean - std
    upper = mean + std

    normal = segment[
        (segment[column_name] >= lower) &
        (segment[column_name] <= upper)
    ]

    abnormal = segment[
        (segment[column_name] < lower) |
        (segment[column_name] > upper)
    ]

    normal_row = {
        "Feed Segment": segment_name,
        "Condition": "Normal",
        "Mean": normal[column_name].mean(),
        "Std": normal[column_name].std(),
        "Variance": normal[column_name].var(),
        "Min": normal[column_name].min(),
        "Max": normal[column_name].max(),
        "Count": len(normal)
    }

    abnormal_row = {
        "Feed Segment": segment_name,
        "Condition": "Abnormal",
        "Mean": abnormal[column_name].mean(),
        "Std": abnormal[column_name].std(),
        "Variance": abnormal[column_name].var(),
        "Min": abnormal[column_name].min(),
        "Max": abnormal[column_name].max(),
        "Count": len(abnormal)
    }

    return [normal_row, abnormal_row]
```

```python
def create_condition(df, col):

    mean = df[col].mean()
    std = df[col].std()

    lower = mean - std
    upper = mean + std

    df = df.copy()

    df["Condition"] = "Normal"

    df.loc[
        (df[col] < lower) |
        (df[col] > upper),
        "Condition"
    ] = "Abnormal"

    return df
```
