```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering

selected_cols = [
    "TOTAL_FEED_RATE",
    "ATTACK AGTR A CURRENTGD2801A-AI.PV",
    "ATTACK AGTR B CURRENTGD2801B-AI.PV",
    "ATTACK AGTR C CURRENTGD2801C-AI.PV",
    "ATTACK AGTR D CURRENTGD2801D-AI.PV"
]

X = df_clean[selected_cols].copy()

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

hc = AgglomerativeClustering(
    n_clusters=2,
    linkage="ward"
)

X["Cluster"] = hc.fit_predict(X_scaled)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))

sns.scatterplot(
    data=X,
    x="TOTAL_FEED_RATE",
    y="ATTACK AGTR B CURRENTGD2801B-AI.PV",
    hue="Cluster",
    palette={0:"green", 1:"red"},
    alpha=0.7
)

plt.title("Hierarchical Clustering - AGTR B")
plt.xlabel("Total Feed Rate")
plt.ylabel("AGTR B Current")

plt.show()
```

```python
def create_condition(df, column):

    plot_df = df.copy()

    mean = plot_df[column].mean()
    std = plot_df[column].std()

    lower_limit = mean - std
    upper_limit = mean + std

    plot_df["Condition"] = "Normal"

    plot_df.loc[
        (plot_df[column] < lower_limit) |
        (plot_df[column] > upper_limit),
        "Condition"
    ] = "Abnormal"

    return plot_df
```

```python
agtr_b = "ATTACK AGTR B CURRENTGD2801B-AI.PV"

plot_df = create_condition(df, agtr_b)
```
